```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.layers import Dense
from keras.models import Sequential
```

```python
df = pd.read_csv(r"C:\Users\KIRPAL SINGH\Desktop\charity_data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.drop(["EIN","NAME"],axis=1)
```

```python
df.head()
```

```python
label_encoder = LabelEncoder()
df["APPLICATION_TYPE"] = label_encoder.fit_transform(df["APPLICATION_TYPE"])
df["AFFILIATION"] = label_encoder.fit_transform(df["AFFILIATION"])
df["CLASSIFICATION"] = label_encoder.fit_transform(df["CLASSIFICATION"])
df["USE_CASE"] = label_encoder.fit_transform(df["USE_CASE"])
df["ORGANIZATION"] = label_encoder.fit_transform(df["ORGANIZATION"])
df["INCOME_AMT"] = label_encoder.fit_transform(df["INCOME_AMT"])
df["SPECIAL_CONSIDERATIONS"] = label_encoder.fit_transform(df["SPECIAL_CONSIDERATIONS"])
```

```python
df.head()
```

```python
for i in range(df.shape[0]):
    df["ASK_AMT"][i] = ((df["ASK_AMT"][i]-df["ASK_AMT"].mean())/df["ASK_AMT"].std())
```

```python
df.head()
```

```python
target = df["IS_SUCCESSFUL"]
features = df.drop(["IS_SUCCESSFUL"],axis=1)
```

```python
df.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.2)
```

```python
x_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(25 , input_shape=(9,) , activation="relu"))
model.add(Dense(500 , activation="relu"))
model.add(Dense(250, activation="relu"))
model.add(Dense(2, activation="sigmoid"))
```

```python
model.compile(optimizer="adam" , loss="sparse_categorical_crossentropy" , metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```
