```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv(r"C:\Users\Owner\Desktop\home work\Module 21\deep-learning-challenge\Resources\charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
application_df = application_df.drop(['EIN','NAME'],axis=1)
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
for i in application_df.columns:
    print(f"the unique value in column {i} is {application_df[i].unique()}")
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

app_counts = application_df['APPLICATION_TYPE'].value_counts()

# determine the cutoff value
cutoff = 1000

# create a list of application types to be replaced
application_types_to_replace = app_counts[app_counts < cutoff].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE

application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
application_df["CLASSIFICATION"].value_counts()[application_df["CLASSIFICATION"].value_counts()>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

clf_counts = application_df['CLASSIFICATION'].value_counts()

# determine the cutoff value
cutoff = 1000

# create a list of application types to be replaced
clf_to_replace = clf_counts[clf_counts < cutoff].index.tolist()


# Replace in dataframe
for cls in clf_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_type_dummies = pd.get_dummies(application_df["APPLICATION_TYPE"])
affiliation_dummies = pd.get_dummies(application_df["AFFILIATION"])
classification_dummies = pd.get_dummies(application_df["CLASSIFICATION"])
usecase_dummies = pd.get_dummies(application_df["USE_CASE"])
organization_dummies = pd.get_dummies(application_df["ORGANIZATION"])
special_considerations_dummies = pd.get_dummies(application_df["SPECIAL_CONSIDERATIONS"])
income_dummies = pd.get_dummies(application_df["INCOME_AMT"])
```

```python
application_df = application_df.drop(["APPLICATION_TYPE","AFFILIATION","CLASSIFICATION","USE_CASE","ORGANIZATION","INCOME_AMT","SPECIAL_CONSIDERATIONS"],axis=1)
application_df = pd.concat([application_df,application_type_dummies,affiliation_dummies,classification_dummies,
                           usecase_dummies,organization_dummies,special_considerations_dummies],axis=1)
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
target = application_df["IS_SUCCESSFUL"]
features = application_df.drop(["IS_SUCCESSFUL"],axis=1)


```

```python
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(features,target,test_size=0.2)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(100 , input_shape=(31,) , activation="relu"))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(500, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(2, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(optimizer="adam" , loss="sparse_categorical_crossentropy" , metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE
history = nn.fit(X_train,Y_train , validation_data=(X_test,Y_test) , epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,Y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
